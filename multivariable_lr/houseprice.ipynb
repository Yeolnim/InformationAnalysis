```python
from statsmodels.formula.api import ols
x=df['Longitude']
y=df['value']
data=pd.DataFrame({'x':x,'y':y})

x_in=data['x'].values.reshape(-1,1)
y_in=data['y'].values.reshape(-1,1)

model=ols('y~x',data).fit()
print(model.summary())
```

```python
import pandas as pd
import statsmodels.api as sm

salary=pd.read_csv('houseprice.csv')

y = salary['value'].values
X = salary[['MedInc','HouseAge','AveRooms','AveOccup','Latitude']]

X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
